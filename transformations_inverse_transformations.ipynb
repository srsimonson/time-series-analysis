# Transformations and Inverse Transformations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import pyplot
from matplotlib.pylab import rcParams
```

```python
rcParams['figure.figsize'] = 15, 6
```

```python
# dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
# df = pd.read_csv('../data_sets/AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)

df = pd.read_csv(
    '../data_sets/AirPassengers.csv',
    parse_dates=['Month'],
    index_col='Month'
)
```

```python
# Convert to time series
ts = df['#Passengers']
```

```python
from datetime import datetime
ts[datetime(1949,1,1)]
```

```python
ts.head()
```

```python
# Log
ts_log = np.log(ts)
print(ts_log.iloc[0])

# Inverse Log = Exponentiation
ts_orig = np.exp(ts_log)
print(ts_orig.iloc[0])

# original
print(ts.iloc[0])
```

```python
# Double Log

ts_double_log = np.log(ts_log)
print(ts_double_log.iloc[0])

# Inverse Double Log
ts_orig2 = np.exp(np.exp(ts_double_log))
print(ts_orig2.iloc[0])
```

```python
# Log Differencing

ts_log_diff = ts_log - ts_log.shift() # difference between the two consecutive log values
print(ts_log_diff.iloc[1])

# Inverse
# add the shifted lag values to the first log value to get the cumulative sum
ts_log_cumsum = ts_log_diff.cumsum()

# Add the cumulative sum to the first log value to get the restored log time series
ts_log_restored = ts_log.iloc[0] + ts_log_cumsum

# Take the exponential of the restored log
ts_orig = np.exp(ts_log_restored)
print(ts_orig)
```

```python
# Log + MA + Differencing
ts_log = np.log(ts)
moving_avg = pd.Series(ts_log).rolling(window=12).mean()
ts_log_moving_avg_diff = ts_log - moving_avg

# Inverse
# Add moving average
ts_log_moving_avg = ts_log_moving_avg_diff + moving_avg

ts_orig = np.exp(ts_log_moving_avg)
print(ts_orig)
```
